```python
#Apertura de archivo csv a un dataframe de pandas

#revisar la ruta para tu caso particular
ruta = "c:\\Users\\jrgui\\Documents\\GitHub\\Machine_Learning\\Reglas_de_Asociacion\\datasets\\Market_Basket_Optimisation.csv"
dataset = pd.read_csv(ruta, header=None)
dataset.head(5)
```

```python
#Creacion de listas de listas con los productos dentro del dataframe

#no se quitan los nan dentro del dataset

transacciones=[]
for i in range(0,7501):
    transacciones.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
#Entrenamiento del algoritmo a priori

from apyori import apriori

reglas = apriori(transactions = transacciones, min_support = 0.003, min_confidence = 0.2, 
                 min_lift = 3, min_length = 2, max_length = 2)

```

```python
resultados = list(reglas)
resultados

# Coloca los resultados organizados en un dataframe de pandas
def inspect(resultados):
    lhs         = [tuple(resultado[2][0][0])[0] for resultado in resultados]
    rhs         = [tuple(resultado[2][0][1])[0] for resultado in resultados]
    supports    = [resultado[1] for resultado in resultados]
    confidences = [resultado[2][0][2] for resultado in resultados]
    lifts       = [resultado[2][0][3] for resultado in resultados]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(resultados), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])

# resultados sin ordenar
resultsinDataFrame

## muestras los resultados de forma descendiente de lift
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
